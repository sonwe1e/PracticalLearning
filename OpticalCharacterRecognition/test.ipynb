```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (20.0, 10.0)
plt.rcParams["figure.dpi"] = 300

# Load image
img = Image.open("./pics/Snipaste_2023-11-05_19-14-25.jpg").convert("L")
img = np.array(img)

# Calculate the horizontal projection of the image
img_h = img.min(axis=1)

# Set a threshold to identify lines
threshold = img_h.max() * 0.8

# plot the horizontal projection and the threshold
plt.plot(img_h)
plt.plot([0, img.shape[0]], [threshold, threshold], "r")
plt.show()
# Find the lines
lines = []
start = 0
end = 0
for i in range(img_h.shape[0]):
    if img_h[i] > threshold:
        start += 1
        if end > start:
            lines.append([start, end])
            start = end
    else:
        end = i

# plot the lines
l = len(lines)
for _, line in enumerate(lines):
    plt.subplot(l, 1, _ + 1)
    plt.axis("off")
    plt.imshow(img[line[0] : line[1], :])
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Load image
img = Image.open("./pics/Snipaste_2023-11-08_10-43-03.jpg").convert("L")
img = np.array(img)

# Calculate the horizontal projection of the image
img_w = img.min(axis=0)

# Set a threshold to identify lines
threshold = img_w.max() * 0.8

# plot the horizontal projection and the threshold
plt.plot(img_w)
plt.plot([0, img.shape[1]], [threshold, threshold], "r")
plt.show()
# Find the lines
lines = []
start = 0
end = 0
for i in range(img_w.shape[0]):
    if img_w[i] > threshold:
        start += 1
        if end > start:
            lines.append([start, end])
            start = end
    else:
        end = i

# plot the lines
l = len(lines)
k = l // 5
for _, line in enumerate(lines):
    plt.subplot(k + 1, 5, _ + 1)
    plt.axis("off")
    plt.imshow(img[:, line[0] : line[1]])
```
